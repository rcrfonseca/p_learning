```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
sns.set(style="whitegrid")

rs = np.random.RandomState(365)
values = rs.randn(365, 4).cumsum(axis=0)
dates = pd.date_range("1 1 2016", periods=365, freq="D")
data = pd.DataFrame(values, dates, columns=["A", "B", "C", "D"])
data = data.rolling(7).mean()

sns.lineplot(data=data, palette="tab10", linewidth=3)
sns.catplot(data=data, kind="bar",palette="tab10", linewidth=3)
```

# seaborn catplot

```python
sns.set(style="ticks")
exercise = sns.load_dataset("exercise")
g = sns.catplot(x="time", y="pulse", hue="kind", data=exercise, legend=False)
plt.legend(loc="upper center",ncol=3)
```

```python
g = sns.catplot(x="time", y="pulse", hue="kind",data=exercise, kind="violin")
```

```python
g = sns.catplot(x="time", y="pulse", hue="kind",col="diet", data=exercise)
```

```python
titanic = sns.load_dataset("titanic")
g = sns.catplot("alive", col="deck", col_wrap=4,
                data=titanic[titanic.deck.notnull()],
                kind="count", 
                height=2.5, aspect=.8)
```

```python
g = sns.catplot(x="age", y="embark_town",
                hue="sex", row="class",
                data=titanic[titanic.embark_town.notnull()],
                orient="h", height=4, aspect=2.5, palette="Set3",
                kind="violin", dodge=True, cut=0, bw=.2)
```
